### Since this dataset is too large we are going to shorten it.
#### As we can see that this dataset is divided on daily basis so now to reduce some of the rows we are going to make this dataset weekly, by taking weekly observations into account only.

```python
li_columns = ["continent", "location", "date", "population", "total_cases", "total_deaths"]
```

```python
print(np.min(df_raw[df_raw.location == "Mexico"]["date"]), np.max(df_raw[df_raw.location == "Mexico"]["date"]))
```

#### We are taking dates from mexico cause it has the maximum range of date collection.

```python
li_dates = []

for i in range(0, df_raw[df_raw.location == "Mexico"].shape[0], 7):
    li_dates.append(df_raw[df_raw.location == "Mexico"]["date"].iloc[i])

print(li_dates)
```

#### These dates are seperated on weekly basis so now we are going to extract the rows only for these dates from every country.

```python
li_data = []
loop = 0

for l in df_raw["location"].unique():
        
        df_temp = df_raw[li_columns][df_raw.location == l]
        
        for d in li_dates:
            if df_temp[df_temp.date == d].shape[0] == 1:
                li_data.append(df_temp[df_temp.date == d].iloc[0])


df_covid = pd.DataFrame(li_data)
df_covid
```

```python
def func_helper1(s):
    
    li_data = []
    loop = 0
    for l in df_covid["location"].unique():
        
        df_temp = df_covid[df_covid.location == l]
        for i in range(1, df_temp.shape[0]):
            
            if df_temp[s].iloc[i] < df_temp[s].iloc[i - 1]:
                li_data.append(df_temp.iloc[i - 1])
                li_data.append(df_temp.iloc[i])
    
    return li_data
```

### Modifying 
#### 1. total_cases 
#### 2. total_deaths 
#### 3. total_tests 
#### 4. total_vaccinations 
### columns according to new dates

```python
df_covid["total_cases"].fillna(0, inplace=True)
df_covid["total_deaths"].fillna(0, inplace=True)

loop = 0
li_cases = []
li_deaths = []

for l in df_covid["location"].unique():
    
    df_temp = df_covid[df_covid.location == l]
    li_cases.append(df_temp["total_cases"].iloc[0])
    li_deaths.append(df_temp["total_deaths"].iloc[0])
    
    for i in range(1, df_temp.shape[0]):
        
        if li_cases[-1] > df_temp["total_cases"].iloc[i]:
            li_cases.append(li_cases[-1])
        else:
            li_cases.append(df_temp["total_cases"].iloc[i])
        
        if li_deaths[-1] > df_temp["total_deaths"].iloc[i]:
            li_deaths.append(li_deaths[-1])
        else:
            li_deaths.append(df_temp["total_deaths"].iloc[i])
    
df_covid["total_cases"] = li_cases
df_covid["total_deaths"] = li_deaths

df_covid
```

#### Adding new columns

```python
new_cases = []
new_deaths = []
loop = 0

for loctn in df_covid["location"].unique():
    
    prev_cases = 0
    prev_deaths = 0
    
    df_temp = df_covid[df_covid.location == loctn]
    
    for i in range(df_temp.shape[0]):
        
        curr_cases = df_temp["total_cases"].iloc[i]
        curr_deaths = df_temp["total_deaths"].iloc[i]
        
        new_cases.append(curr_cases - prev_cases)
        new_deaths.append(curr_deaths - prev_deaths)
        
        prev_cases = curr_cases
        prev_deaths = curr_deaths
        
df_covid["new_cases"] = new_cases
df_covid["new_deaths"] = new_deaths

df_covid
```
